```python
import pandas as pd
import matplotlib.pyplot as plt

# Load COVID-19 testing data
covid_data = pd.read_csv("covid.csv", parse_dates=["test_date"])

# Load student enrollment data for 2019 and 2020
enrollment_2019 = pd.read_csv("student_enrollment_2019.csv", skiprows=9, index_col=0)
enrollment_2020 = pd.read_csv("student_enrollment_2020.csv", skiprows=9, index_col=0)

# Data Cleaning
enrollment_2019 = enrollment_2019.iloc[:-1]  # Remove "Total all students" row
enrollment_2020 = enrollment_2020.iloc[:-1]  # Remove "Total all students" row
covid_data["age_group"] = covid_data["age_group"].str.extract(r'(\d+)').astype(float)

# Data Exploration and Analysis
# Calculate percentage change in total enrollment
total_enrollment_2019 = enrollment_2019.sum(axis=1)
total_enrollment_2020 = enrollment_2020.sum(axis=1)
enrollment_change_2019 = ((total_enrollment_2019 - total_enrollment_2019) / total_enrollment_2019) * 100
enrollment_change_2020 = ((total_enrollment_2020 - total_enrollment_2019) / total_enrollment_2019) * 100

# Visualization
plt.figure(figsize=(12, 8))

# Plot 2019 Enrollment Change
plt.subplot(2, 1, 1)
plt.plot(enrollment_change_2019.index, enrollment_change_2019.values, marker="o", label="2019")
plt.xlabel("Grade/Year")
plt.ylabel("Percentage Change in Enrollment")
plt.title("Percentage Change in Student Enrollment (2019)")
plt.xticks(rotation=45)
plt.legend()

# Plot 2020 Enrollment Change
plt.subplot(2, 1, 2)
plt.plot(enrollment_change_2020.index, enrollment_change_2020.values, marker="o", label="2020", color="orange")
plt.xlabel("Grade/Year")
plt.ylabel("Percentage Change in Enrollment")
plt.title("Percentage Change in Student Enrollment (2020)")
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()

# Combined Plot for 2019 and 2020 Enrollment Changes
plt.figure(figsize=(12, 6))
plt.plot(enrollment_change_2019.index, enrollment_change_2019.values, marker="o", label="2019")
plt.plot(enrollment_change_2020.index, enrollment_change_2020.values, marker="o", label="2020", color="orange")
plt.xlabel("Grade/Year")
plt.ylabel("Percentage Change in Enrollment")
plt.title("Percentage Change in Student Enrollment (2019-2020)")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
