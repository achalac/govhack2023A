```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('Toilet_Accessibility.csv')

# Get user input for the desired location
desired_location = input("Enter the location you're interested in: ")

# Filter the DataFrame based on the desired location
filtered_data = df[df['LOCATION_NAME'] == desired_location]

# Check if any data matches the user's input
if not filtered_data.empty:
    access_activity_level = filtered_data['ACCESS_ACTIVITY_LEVEL'].iloc[0]
    print(f"ACCESS_ACTIVITY_LEVEL for {desired_location}: {access_activity_level}")
else:
    print(f"No data found for {desired_location}.")


```

```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('Toilet_Accessibility.csv')

# Filter the DataFrame to extract rows with 'ACCESS_ACTIVITY_LEVEL' equal to 'high'
high_access_toilets = df[df['ACCESS_ACTIVITY_LEVEL'] == 'High']

# Extract 'HOURS_OPEN_PER_DAY' and 'LOCATION_NAME' columns
selected_columns = high_access_toilets[['HOURS_OPEN_PER_DAY', 'LOCATION_NAME']]

# Display the extracted data with "High Access" indication
for index, row in selected_columns.iterrows():
    print(f"Location: {row['LOCATION_NAME']}")
    if row['HOURS_OPEN_PER_DAY']:
        print(f"Hours Open per Day: {row['HOURS_OPEN_PER_DAY']}")
    print("High Access. Has Wheelchair Access\n")
```
