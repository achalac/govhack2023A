```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('bblocations.csv')

# Group the data by 'PARK_NAME' and sum the 'NUMBER_OF_PLATES' for each park
park_plate_counts = df.groupby('PARK_NAME')['NUMBER_OF_PLATES'].sum()

# Create a bar plot
plt.figure(figsize=(10, 6))
park_plate_counts.plot(kind='bar')
plt.title('Number of Plates in Each Park')
plt.xlabel('Park Name')
plt.ylabel('Number of Plates')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Display the plot
plt.show()
```

```python
#Pie Chart: Proportion of Plates per Park

import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('bblocations.csv')

# Group the data by 'PARK_NAME' and sum the 'NUMBER_OF_PLATES' for each park
park_plate_counts = df.groupby('PARK_NAME')['NUMBER_OF_PLATES'].sum()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(park_plate_counts, labels=park_plate_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Plates per Park')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('bblocations.csv')

# Get user input for the desired suburb
desired_suburb = input("Enter the suburb you're interested in: ")

# Filter the DataFrame based on the desired suburb
filtered_data = df[df['SUBURB'] == desired_suburb]

# Display the BBQ locations and number of plates in the desired suburb
if not filtered_data.empty:
    print(f"BBQ Locations and Plates in {desired_suburb}:\n")
    for index, row in filtered_data.iterrows():
        print(f"Location: {row['SUBURB']}")
        print(f"Number of Plates: {row['NUMBER_OF_PLATES']}\n")
else:
    print(f"No BBQ Locations found in {desired_suburb}.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('bblocations.csv')

# Group the data by 'PARK_NAME' and sum the 'NUMBER_OF_PLATES' for each park
park_plate_counts = df.groupby('PARK_NAME')['NUMBER_OF_PLATES'].sum()

# Create a bar plot
plt.figure(figsize=(10, 6))
park_plate_counts.plot(kind='bar')
plt.title('Number of Plates in Each Park')
plt.xlabel('Park Name')
plt.ylabel('Number of Plates')

# Customize x-axis labels
plt.xticks(rotation=45, ha='right')
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(10))  # Display a maximum of 10 x-axis labels

plt.tight_layout()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('bblocations.csv')

# Group the data by 'PARK_NAME' and sum the 'NUMBER_OF_PLATES' for each park
park_plate_counts = df.groupby('PARK_NAME')['NUMBER_OF_PLATES'].sum()

# Create a pie chart
plt.figure(figsize=(8, 8))
# Plot only a subset of the data for the pie chart (top 10 parks)
top_parks = park_plate_counts.nlargest(10)
plt.pie(top_parks, labels=top_parks.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Plates per Top 10 Parks')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the legend
plt.legend(top_parks.index, loc="best")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('bblocations.csv')

# Group the data by 'PARK_NAME' and sum the 'NUMBER_OF_PLATES' for each park
park_plate_counts = df.groupby('PARK_NAME')['NUMBER_OF_PLATES'].sum()

# Create a pie chart
plt.figure(figsize=(8, 8))
# Plot only a subset of the data for the pie chart (top 10 parks)
top_parks = park_plate_counts.nlargest(10)
plt.pie(top_parks, labels=top_parks.index, autopct=lambda p: '{:.0f}'.format(p * sum(top_parks) / 100))
plt.title('Proportion of Plates per Top 10 Parks')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the legend
plt.legend(top_parks.index, loc="best")

plt.show()
```
